# Multimodal starter — price series + candlestick chart

Candlestick PNGs are rendered on demand via `src/tdwm/charts.py`. They
are NOT stored in the HF dataset by default because they are largely
redundant with OHLCV — turn them on only if your model needs them.

```python
from datasets import load_dataset

ds = load_dataset('twelvedata/financial-world-model', 'bars_1day', split='train')
df = ds.to_pandas()
aapl = df[df['symbol']=='AAPL'].tail(60).copy()
aapl.tail()
```

```python
from pathlib import Path
import sys
sys.path.insert(0, '../src')
from tdwm.charts import render_candlestick_png
render_candlestick_png(aapl, Path('aapl_last60.png'), title='AAPL last 60 bars')
```

From here you can feed (series, image) pairs into a
vision-language adapter. Be careful to align the image timestamp with
the `as_of` of the numeric features — i.e. never include the next-day
bar in the image if the task is next-bar prediction.
